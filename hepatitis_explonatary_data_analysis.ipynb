```python
data_modification = {'False': 0,
               'True': 1,
               'die': 0,
               'live': 1,
               '?': np.nan,
               'female': 0,
               'male': 1}


hepatitis.replace(data_modification, inplace = True)
```

```python
hepatitis['steroid'] = hepatitis['steroid'].replace(['FALSE'],'0')
hepatitis['steroid'] = hepatitis['steroid'].replace(['TRUE'],'1')
```

```python
hepa_drop = hepatitis.dropna()
```

```python
total_of_patients = hepatitis.shape[0]
total_of_live_patients = (np.sum(hepatitis['class'] == 1)/total_of_patients)*100
total_of_dead_patients = (np.sum(hepatitis['class'] == 0)/total_of_patients)*100
print("Living patients:", round(total_of_live_patients,2),"%")
print("Dead patients:", round(total_of_dead_patients,2),"%")
```
